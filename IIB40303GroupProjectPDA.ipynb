```python
conditions = [
    (df['Quality of Sleep'] <= 4),
    (df['Quality of Sleep'] <= 7),
    (df['Quality of Sleep'] > 7)
]

values = ['Poor', 'Moderate', 'Good']

df['Sleep Quality Category'] = np.select(conditions, values, default='Unknown')

sns.countplot(x='Sleep Quality Category', data=df)
plt.title('Sleep Quality Categories')
plt.show()

#Creates rules/categories.

#<= 4
#Poor sleep.

#<= 7
#Moderate sleep.

#> 7
#Good sleep.
```
